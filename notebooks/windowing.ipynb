```python
#GABOR
import numpy as np
from scipy import ndimage as ndi
from skimage.filters import gabor_kernel
from math import pi, sqrt

def apply_gabor_filter(image, frequency, theta, sigma):
    image = image.astype(np.float32)
    image = (image - image.mean()) / (image.std() + 1e-8)
    kernel = gabor_kernel(frequency, theta=theta, sigma_x=sigma, sigma_y=sigma)
    real = ndi.convolve(image, np.real(kernel), mode='wrap')
    imag = ndi.convolve(image, np.imag(kernel), mode='wrap')
    return np.sqrt(real**2 + imag**2)

def get_all_filtered(image):
    all_filtered = []
    for frequency in [1/8, 1/4]:
        sigma = 1/(2*frequency)
        #for theta in [0, pi/6, pi/3, pi/2, 2*pi/3, 5*pi/6]:
        for theta in [pi/6, 2*pi/3]:
            filtered = apply_gabor_filter(image, frequency, theta, sigma)
            all_filtered.append(filtered)
    return all_filtered

def spatial_entropy_uniformity(tile, grid_size=(32, 32)):
    h, w = tile.shape
    gh, gw = grid_size
    patch_entropies = []

    for i in range(gh):
        for j in range(gw):
            patch = tile[
                i * h // gh : (i + 1) * h // gh,
                j * w // gw : (j + 1) * w // gw
            ]
            patch_entropies.append(shannon_entropy(patch))
    
    patch_entropies = np.array(patch_entropies)
    mean_entropy = patch_entropies.mean()
    entropy_variance_penalty = patch_entropies.std()
    
    # Reward high average entropy, penalize uneven entropy spread
    score = mean_entropy - entropy_variance_penalty
    return score

def window(image, lower_bound, upper_bound):
    image = np.asarray(image)
    image = np.clip(image, lower_bound, upper_bound)
    image = ((image - lower_bound) / (upper_bound - lower_bound) * 255).astype(np.uint8)
    return image

def mutual_information(image_12bit, image_8bit, bins_12=4096, bins_8=256):
    joint_hist, _, _ = np.histogram2d(
        image_12bit.ravel(),
        image_8bit.ravel(),
        bins=(bins_12, bins_8)
    )
    joint_prob = joint_hist / joint_hist.sum()

    p_I = np.sum(joint_prob, axis=1)
    p_I_tilde = np.sum(joint_prob, axis=0)

    def entropy(p):
        p = p[p > 0]
        return -np.sum(p * np.log2(p))

    H_I = entropy(p_I)
    H_I_tilde = entropy(p_I_tilde)
    H_joint = entropy(joint_prob)

    mi = H_I + H_I_tilde - H_joint
    return mi, joint_hist, p_I, p_I_tilde


def evaluate(orig, orig_filtered, a, b):
    windowed = window(orig, a, b)
    windowed_filtered = get_all_filtered(windowed)
    mis = []
    for single_orig_filtered, single_windowed_filtered in zip(orig_filtered, windowed_filtered):
        mi, joint_hist, p_I, p_I_tilde = mutual_information(single_orig_filtered, single_windowed_filtered)
        mis.append(mi)
    return np.average(mis)
```

```python
import numpy as np

import numpy as np
from skimage.transform import resize

def largest_square_in_mask(image: np.ndarray, mask: np.ndarray, downscale: int = 16) -> np.ndarray:
    if image.shape[:2] != mask.shape:
        raise ValueError("Image and mask dimensions do not match.")

    h, w = mask.shape
    if downscale > 1:
        small_mask = resize(mask, (h // downscale, w // downscale), order=0, preserve_range=True, anti_aliasing=False).astype(mask.dtype)
    else:
        small_mask = mask

    small_mask = small_mask != 0
    sh, sw = small_mask.shape
    dp = np.zeros((sh, sw), dtype=np.int32)

    max_size = 0
    max_pos = (0, 0)

    for i in range(sh):
        for j in range(sw):
            if small_mask[i, j]:
                if i == 0 or j == 0:
                    dp[i, j] = 1
                else:
                    dp[i, j] = min(dp[i-1, j], dp[i, j-1], dp[i-1, j-1]) + 1
                if dp[i, j] > max_size:
                    max_size = dp[i, j]
                    max_pos = (i, j)

    if max_size == 0:
        return None

    i, j = max_pos
    y1 = (i - max_size + 1) * downscale
    x1 = (j - max_size + 1) * downscale
    size = max_size * downscale

    y2 = min(y1 + size, h)
    x2 = min(x1 + size, w)

    y1 = max(0, y2 - size)
    x1 = max(0, x2 - size)

    return image[y1:y2, x1:x2]
```

```python
import time

dataset = MammoDataset(DatasetEnum.RSNA, Split.ALL, return_mode=ReturnMode.DICOM)

for dcm in dataset:
    img = dcm.pixel_array
    img = negate_if_should(img)
    img = flip_if_should(img)
    mask = get_dicom_mask(img)
    
    orig = img
    
    img = largest_square_in_mask(img, mask)
    
    
    img = resize_img(img, (256, 256))
    
#     plt.imshow(img)
#     plt.show()
    
    all_filtered_img = get_all_filtered(img)
    
    max_val = img.max()
    step = max_val//10
    
    start = time.time()
    
    a_values = list(range(0, max_val, step))
    b_values = list(range(0, max_val, step))
    heatmap = []
    
    best_a = 0
    best_b = 0
    best_val = 0
    
    for i, a in enumerate(range(0, max_val, step)):
        row = []
        for j, b in enumerate(range(0, max_val, step)):
            if b > a:
                val = evaluate(img, all_filtered_img, a, b)
                if val > best_val:
                    best_a = a
                    best_b = b
                    best_val = val
            else:
                val = np.nan
            row.append(val)
        heatmap.append(row)
        
    print(img.max())
    
    plt.imshow(orig, cmap='gray')
    plt.show()
    plt.imshow(window(orig, best_a - np.uint16(0.3 * best_a), best_b + (0.3 * np.uint16(max_val - best_b))), cmap='gray')
    plt.show()
    plt.imshow(window(orig, best_a, best_b), cmap='gray')
    plt.show()
    plt.imshow(window(orig, best_a + np.uint16(0.3 * best_a), best_b - np.uint16(0.3 * (max_val - best_b))), cmap='gray')
    plt.show()
        
#     heatmap = np.array(heatmap)

#     fig, ax = plt.subplots()
#     cax = ax.imshow(heatmap, origin='lower', interpolation='nearest', extent=[0, max_val, 0, max_val])
#     ax.set_xlabel('b')
#     ax.set_ylabel('a')
#     fig.colorbar(cax, label='Evaluation Value')
#     plt.title('Heatmap of Evaluation over (a, b)')
#     plt.show()
        
    end = time.time()
    print("Elapsed time:", end - start, "seconds")

```

```python
def show_mammo(cc, mlo, tiles):
    def resize(img):
        return cv2.resize(img, (256, 256), interpolation=cv2.INTER_AREA)

    display_tiles = []
    titles = []

    if cc is not None:
        display_tiles.append(resize(cc))
        titles.append('CC')
    if mlo is not None:
        display_tiles.append(resize(mlo))
        titles.append('MLO')

    display_tiles.extend([resize(tile) for tile in tiles])
    titles.extend([''] * len(tiles))

    total = len(display_tiles)
    cols = 10
    rows = (total + cols - 1) // cols

    fig = plt.figure(figsize=(cols, rows))

    for i, (img, title) in enumerate(zip(display_tiles, titles)):
        ax = fig.add_subplot(rows, cols, i + 1)
        ax.imshow(img, cmap='gray')
        if title:
            ax.set_title(title)
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```
